```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1
hbar = 1
a = 1
N=1000
x = np.linspace(0, a, N + 1)
dx = x[1] - x[0]
M = np.diag(-2 * np.ones(N - 1)) + np.diag(np.ones(N - 2), 1) + np.diag(np.ones(N - 2), -1)
print(M)
print(M.shape)
```

```python
k= -hbar ** 2 / (2 * m * (dx ** 2))
H = k * M
E, psi = np.linalg.eigh(H)
```

```python
print(psi.T[0])
```

```python
xp = np.linspace(dx, a - dx, N - 1)
```

```python
plt.plot(xp, psi.T[1])
plt.grid(True)
plt.xlabel("x")
plt.ylabel("psi")
plt.show()
```

```python
# check normalization
psi2 = psi.T[1]
I = np.sum(psi2 ** 2 * dx)
print(I)
A = 1 / np.sqrt(I)
psi2 = A * psi2
I = np.sum(psi2 ** 2 * dx)
print(I)
```

```python
plt.plot(xp, psi2 ** 2)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("psi*psi")
plt.show()
```
